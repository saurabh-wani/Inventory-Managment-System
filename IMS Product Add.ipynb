```python
import json
import time
```

```python
fd = open("Records.json","r")
read = fd.read()
fd.close()

records = json.loads(read)
```

```python
records
```

# Show Records Before addition of old or new Product

```python
# show details of records

print("Product Id   ","Price     ","      Quantity       ","    Product" )
for i in records:
    print(i,"          " , records[i]['Price( in RS.)'],"           ", records[i]['Quantity_Available'],"                ",records[i]['Product_Name'],)
```

```python
prod_lst = []
for i in records:
    prod_lst.append(i)
    
prod_lst
```

# Add old or new products to inventory

```python
while(input("Y/N: Do you want to add another product:").lower() == 'y'):

    prod_id = str(input("Enter Product id: "))



    if(prod_id in prod_lst):
        print("Y/N:  Do you want to add quantity of ")
        if(input(records[prod_id]['Product_Name']).lower() == 'y'):
            new = int(input("Enter no. of quantity You want to add: " ))
            records[prod_id]['Quantity_Available'] += new
            print("You have added",new,records[prod_id]['Product_Name'],"s'  to inventory." )
        else:
            print(prod_id," This product Id is already assigned to",records[prod_id]['Product_Name'])
            print("Please enter valid product id")
    else:
        product_name = str(input("Enter Product name: "))
        product_company = str(input("Enter Product's Company name: "))
        quantity = int(input("Enter Quantity: "))
        price = int(input("Enter Price: "))
        weight = int(input("Enter weight of product in grams: "))

        records[prod_id] = {'Product_Name': product_name, 'Product_Company': product_company, 'Quantity_Available': quantity,'Price( in RS.)': price,'Weight( in gms)':weight }

        
    #update records in json file
    
    js = json.dumps(records)

    fd = open("records.json",'w')
    fd.write(js)
    fd.close()
```

# Delete any existing product records

```python
#for deleting entire product record
#use carefully otherwise entire details of particular product will be deleted


while(input("Y/N: Do you want to delete another product:").lower() == 'y'):
    
    delete = input("Please Enter Product Id of product you want to delete from Inventory Managment System:")
    print("Do you want to delete all details of",records[delete]['Product_Name']," Y/N: ")
    
    if(input("Y or N".lower() == 'y')):
        print("You are deleting records of ",records[delete]['Product_Name'],".......................")
        del records[delete]
        print("All details of specified product have been deleted")
        
    else:
        print("You haven't deleted records of",records[delete]['Product_Name'])
    
    #update records in json file
    
    js = json.dumps(records)

    fd = open("records.json",'w')
    fd.write(js)
    fd.close()
```
